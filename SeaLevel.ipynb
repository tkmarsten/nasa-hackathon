```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Import data, create dataframe
df = pd.read_csv('sea_levels.csv', header=None, skiprows=50, usecols=[2,11])
df.head()
```

```python
df.shape
```

```python
df.columns = ['years', 'sea level']
df.head()
```

```python
# Plot
years = df.iloc[:, 0]
sea_level = df.iloc[:, 1]

plt.scatter(years, sea_level)
plt.title('Scatter Plot of Global Mean Sea Level vs Year')
plt.xlabel('Years', fontsize=12)
plt.ylabel('Global Mean Sea Level', fontsize=12)
plt.show()
```

```python
# Graph begins ~0 mm 
sea_level = sea_level + abs(np.min(sea_level))
```

```python
# Create a linear regression from the data points
m, b = np.polyfit(years, sea_level, 1)

def f(x):
    return m*x + b

plt.scatter(years, sea_level)
plt.plot(years, f(years), "r--")
plt.title('Scatter Plot of Global Mean Sea Level vs Year')
plt.xlabel('Years', fontsize=12)
plt.ylabel('Global Mean Sea Level (mm)', fontsize=12)
plt.show()

print('y = {0} * x + {1}'.format(m, b))
```

```python
# Years column (2D array)
x = df['years'].values
# Sea level column (2D array)
y = df['sea level'].values

# Pick the Linear Regression model and instantiate it
model = LinearRegression(fit_intercept=True)

# Fit the model
model.fit(x.reshape(-1, 1), y)
# Predict using the linear model
mean_predicted = model.predict(x[:, np.newaxis])

# Plot
plt.scatter(years, sea_level)
plt.plot(years, mean_predicted, 'r--')
plt.title('Scatter Plot of Global Mean Sea Level vs Year')
plt.xlabel('Years', fontsize=12)
plt.ylabel('Global Mean Sea Level (mm)', fontsize=12)
plt.show()

print("coefficient = {0}, and intercept = {1}".format(model.coef_[0], 
      model.intercept_))
```

```python
# Mean of sea_level column
mean_sea_level = np.mean(sea_level)

# Range of sea level column
range_sea_level = np.max(sea_level) - np.min(sea_level)

# Mean normalization of sea level column
mean_norm_sea_level = (sea_level - mean_sea_level) / range_sea_level
```

```python
# Convert to 2D array
mean_norm_y = mean_norm_sea_level.values

# Plot
plt.scatter(years, mean_norm_sea_level)
plt.title('Mean Normalized Sea Levels vs Years')
plt.xlabel('Years', fontsize=12)
plt.ylabel('Mean Normalized Sea Levels (mm)', fontsize=12)
plt.show()
```
